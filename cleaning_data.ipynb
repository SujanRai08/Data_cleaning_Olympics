# What should we clean up ?

- [ ] remove bullet points in the used name
- [ ] extra columns with nan values
- [ ] split born data and address
- [ ] died in data and address
- [ ] measurements removing height and weight
- [ ] remove petnames

```python
df = bios.copy()
```

```python
df.info()
```

```python
df['name'] = df['Used name'].str.replace('•',' ')
```

```python
df.head()
```

```python
df[['height_cm','weight_kg']] = df['Measurements'].str.split('/', expand=True)
```

```python
df['height_cm'].unique()
```

```python
def correct_units(values):
    if isinstance(values,str) and 'kg' in values:
        return values.replace('kg','cm')
    return values
```

```python
df['height_cm'] = df['height_cm'].apply(correct_units)
```

```python
import re
def process_ranges(value):
    if isinstance(value,str) and '-' in value:
        # Extract numbers, calculate average, and append 'cm'
        numbers = list(map(int,re.findall(r'\d+',value)))
        return f'{sum(numbers)/ len(numbers):.1f} cm'
    return value

df['height_cm'] = df['height_cm'].apply(process_ranges)
```

```python
def clean_height(value):
    if pd.isna(value): # handle nan values
        return None
    value = value.strip() #remove extra spaces
    if 'kg' in value:
        value = value.replace('kg','cm')
    if ',' in value:
        numbers = list(map(int,re.findall(r'\d+',value)))
        average  = sum(numbers) / len(numbers)
        return f'{average:.1f}cm'
    elif '-' in value:
        numbers = list(map(int,re.findall(r'\d+',value)))
        average =sum(numbers)/len(numbers)
        return f"{average:.1f} cm"
    elif isinstance(value, str) and re.search(r'\d+cm', value):  # Ensure space before 'cm'
        return add_space_to_unit(value)
    
    elif re.search(r'\d+ cm',value):
        return value
    else:
        return None
    
def add_space_to_unit(value):
    if isinstance(value,str):
        return re.sub(r'(?<=\d)cm',' cm',value)
    return value
    
df['height_cm'] = df['height_cm'].apply(clean_height)
```

```python
df['weight_kg'].unique()
```

```python
def clean_weight(value):
    if pd.isna(value): 
        return None
    value = value.strip() 
    if 'cm' in value:  
        value = value.replace('cm', 'kg')
    if ',' in value or '-' in value: 
        numbers = list(map(int, re.findall(r'\d+', value)))
        if numbers:  
            average = sum(numbers) / len(numbers)
            return f"{average:.1f} kg"  
        return None
    elif isinstance(value, str) and re.search(r'\d+kg$', value):  
        return add_space_to_kg(value)
    elif isinstance(value, str) and re.search(r'\d+ kg$', value):
        return value
    else:
        return None  

def add_space_to_kg(value):
    if isinstance(value, str):
        return re.sub(r'(?<=\d)kg$', ' kg', value)
    return value 


df['weight_kg'] = df['weight_kg'].apply(clean_weight)
```

```python
df['weight_kg']
```

```python
df.head()
```

```python
df['height_cm'] = pd.to_numeric(df['height_cm'].str.replace(' cm','',regex=False),errors='coerce')
df['weight_kg'] = pd.to_numeric(df['weight_kg'].str.replace(' kg','',regex=False),errors='coerce')
```

```python
df['height_cm'].dtype
df['weight_kg'].dtype
print(df['height_cm'].dtype)
print(df['weight_kg'].dtype)
```

```python
df['height_cm'] = df['height_cm'].apply(lambda x: int(x) if pd.notna(x) else pd.NA)
df['weight_kg'] = df['weight_kg'].apply(lambda x: int(x) if pd.notna(x) else pd.NA)
```

```python
# Ensure columns are of nullable integer type
df['height_cm'] = df['height_cm'].astype('Int64')
df['weight_kg'] = df['weight_kg'].astype('Int64')
```

```python
df.drop(columns=['Measurements'],inplace=True)
```

```python
df.drop(columns=['Used name'],inplace=True)
```

```python
df = df.sort_values(by=['height_cm', 'weight_kg'], ascending=[True, False])
```

```python
date_pattern = r'(\d+ \w+ \d{4}|\d{4})'
df['born_date'] = df['Born'].str.extract(date_pattern)
df['born_year'] = df['Born'].str.extract(r'(\d{4})') 
```

```python
df['born_date'] = pd.to_datetime(df['born_date'], format="mixed", errors='coerce')
df['born_year'] = pd.to_numeric(df['born_year'])
```

```python
df['died_date'] = df['Died'].str.extract(date_pattern)
df['died_date'] = pd.to_datetime(df['died_date'], format="mixed", errors='coerce')
```

```python
# get city, region and country from born column
location = r'in ([\w\s()-]+),([\w\s-]+) \((\w+)\)'
df['Born'].str.extract(location)
```

```python
df[['born_city','born_region','born_country']] = df['Born'].str.extract(location,expand=True)
```
